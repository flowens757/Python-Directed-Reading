```python
###Create initial bot

import random
#Create a therapist Bot
random_responses = ["That is quite interesting, please tell me more.",
                    "I see. Do go on.",
                    "Why do you say that?",
                    "Know your worth.",
                    "Do you believe that?",
                    "Why do you feel that way?",
                   "Who hurt you?"]

print("Hello, I am Kevin, your therapist.")
print("If you don't like me, you can always get a new therapist by saying 'bye'")
print("How are you today?")

while True:
    # wait for the user to enter some text
    user_input = input("> ")
    if user_input.lower() == "bye":
        # if they typed in 'bye' (or even BYE, ByE, byE etc.), break out of the loop
        break
    else:
        response = random.choices(random_responses)[0]
    print(response)

print("It was nice talking to you, goodbye!")
```

```python
#Make a better therapy bot that can actually talk and probes for elaboration 
import random
#Build a set of responses
keyword_responses = {
    "sad": ["I'm sorry to hear that. What's been making you feel sad?", "It's tough when we feel sad. Can you tell me more about it?"],
    "anxious":  ["Anxiety can be overwhelming. Take a deep breath and try to focus on the present moment.", "I understand that feeling anxious can be challenging. Is there something specific that's causing your anxiety?"],
    "nervous": ["Anxiety can be overwhelming. Take a deep breath and try to focus on the present moment.", "I understand that feeling anxious can be challenging. Is there something specific that's causing your anxiety?"],
    "stressed": ["Stress can take a toll on our well-being. How can I support you in managing your stress?", "It's important to find healthy ways to cope with stress. What do you typically do when you feel stressed?"],
    "depressed": ["Life can be incredibly hard. What do you think is causing you to feel this way?", "How long have you been feeling this way?"],
    "struggling": ["You may need to focus on self-compassion more. I'd listen to this podcast https://self-compassion.org", "You do not have to struggle in silence. Please tell me more."],
    "advice": ["I'm here to help you discover yourself and navigate these challenging times.", "Let's talk more, in the meantime here is a helpful link: https://positivepsychology.com/self-compassion-exercises-worksheets/"],
    "happy": ["Let's focus on those emotions. What are some things that make you happy?", "It's important to remember that your feelings are valid. Can you tell me a little more about this?"],
    "help": ["How can I help you?"], 
}


#Start the converstaion
print("Stefanie (Therapist): Hi, I'm here to help. How can I assist you today? If you'd prefer not to talk to me, just tell me 'bye'.")
print("Stefanie (Therapist): Feel free to share anything that's been on your mind. Remember it is ok to not be ok.")

while True:
    #User response
    user_input = input("Me: ")

    #If user wants to end conversation
    if user_input.lower() in ["bye", "goodbye", "exit"]:
        print("Stefanie (Therapist): Thank you for sharing. Take care!")
        break

    #Check for key words
    matched_keywords = []
    for keyword in keyword_responses:
        if keyword in user_input.lower():
            matched_keywords.append(keyword)

    # Generate a response
    if matched_keywords:
        response = random.choice(keyword_responses[random.choice(matched_keywords)])
    else:
        response = random.choice(["Why do you think that?", "I see. Tell me more.", "Ok, let's talk more about that.", "Could you elaborate?"])

    print("Stefanie (Therapist):", response)
```
