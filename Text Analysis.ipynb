```python
#Find the longest updates
import nltk
nltk.download('punkt')

def get_longest_update(status_updates, length_type="tokens"):
    longest_update = ""
    max_length = 0

    for update in status_updates:
        message = update.get("status_message", "")
        if length_type == "tokens":
            tokens = nltk.word_tokenize(message)
            length = len(tokens)
        elif length_type == "sentences":
            sentences = nltk.sent_tokenize(message)
            length = len(sentences)
        elif length_type == "characters":
            length = len(message)
        else:
            raise ValueError("Invalid length_type parameter. Choose 'tokens', 'sentences', or 'characters'.")

        if length > max_length:
            max_length = length
            longest_update = message

    return longest_update
#Call the status updates
status_updates = [
    {'link_name': 'Timeline Photos', 'num_angrys': '7', 'num_comments': '543', 'num_hahas': '17', 'num_likes': '6178', 'num_loves': '572', 'num_reactions': '6813', 'num_sads': '0', 'num_shares': '359', 'num_wows': '39', 'status_id': '153080620724_10157915294545725', 'status_link': 'https://www.facebook.com/DonaldTrump/photos/a.488852220724.393301.153080620724/10157915294545725/?type=3', 'status_message': 'Beautiful evening in Wisconsin- THANK YOU for your incredible support tonight! Everyone get out on November 8th - and VOTE! LETS MAKE AMERICA GREAT AGAIN! -DJT', 'status_published': '10/17/2016 20:56:51', 'status_type': 'photo'},
    {'link_name': '', 'num_angrys': '5211', 'num_comments': '3644', 'num_hahas': '75', 'num_likes': '26649', 'num_loves': '487', 'num_reactions': '33768', 'num_sads': '191', 'num_shares': '17653', 'num_wows': '1155', 'status_id': '153080620724_10157914483265725', 'status_link': 'https://www.facebook.com/DonaldTrump/videos/10157914483265725/', 'status_message': "The State Department's quid pro quo scheme proves how CORRUPT our system is. Attempting to protect Crooked Hillary, NOT our American service members or national security information, is absolutely DISGRACEFUL. The American people deserve so much better. On November 8th, we will END this RIGGED system once and for all!", 'status_published': '10/17/2016 18:00:41', 'status_type': 'video'}
]

#Longest message by tokens
longest_message = get_longest_update(status_updates, length_type="tokens")
print("Longest update (by tokens):")
print(longest_message)

#Longest message by sentence length
longest_message = get_longest_update(status_updates, length_type="sentences")
print("Longest update (by sentences):")
print(longest_message)

#Longest message by characters
longest_message = get_longest_update(status_updates, length_type="characters")
print("Longest update (by characters):")
print(longest_message)
```

```python
#Find updates that mention Obama or Clinton
def get_updates_with_keywords(status_updates, keywords, case_sensitive=False):
    filtered_status_updates = []
    for update in status_updates:
        message = update.get("status_message", "")
        if not case_sensitive:
            message = message.lower()
            keywords = [keyword.lower() for keyword in keywords]
        tokens = nltk.word_tokenize(message)
        if any(keyword in tokens for keyword in keywords):
            filtered_status_updates.append(update)
    return filtered_status_updates

#Test --> Some issue with Trump tweets data but it returns a few tweets
keywords = ["Clinton", "Obama"]
filtered_updates = get_updates_with_keywords(status_updates, keywords)
print("Status updates with keywords:")
for update in filtered_updates:
    print(update.get("status_message"))
```

```python
def get_laureates(dict_prizes, year=None, category=None):
    prizes = dict_prizes.get("prizes", [])
    filtered_prizes = [
        prize
        for prize in prizes
        if (year is None or prize.get("year") == str(year))
        and (category is None or prize.get("category") == category)
    ]
    laureates = [
        laureate["firstname"] + " " + laureate["surname"]
        for prize in filtered_prizes
        for laureate in prize.get("laureates", [])
    ]
    return laureates
```
